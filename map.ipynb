```python
df3 = pd.read_csv('tables\\df3.csv', low_memory = False)[['beat', 'trr_year', 'side', 'point']]
df3 = df3.groupby(['trr_year', 'beat', 'side'])['point'].agg(pd.Series.mode)
df3.shape
```

```python
df0 = pd.read_csv('tables/data_area_poly.csv').drop(columns = ['Unnamed: 0'])
df0 = df0.rename(columns = {'name': 'beat'})
df0.head(2)
```

```python
df = pd.merge(df0, df3, left_on=['beat', 'side'], right_on=['beat', 'side'])
print(df.shape)
df.head(2)
```

```python
df['point'][0][0]
```

```python
import re
import numpy as np

points = df['point'].apply( lambda x: re.findall(r'(-?[1-9]+.?[1-9]+)', x))
lat = points.apply( lambda x: eval(x[0]) )
long = points.apply( lambda x: eval(x[1]) )
```

```python
lat
```

```python
import json
area = df['st_asgeojson'].apply(lambda x: json.loads(x))
area[0]
```

```python
import geopandas as gpd

fig = px.choropleth(df, geojson=area, lat = lat, lon = long, color='side',
                    color_continuous_scale='Viridis',
                    scope='usa',
                    labels={'side': 'side'}
                    )
# fig.update_geos(fitbounds="locations")
fig.update_layout(margin={'r': 0, 't': 0, 'l': 0, 'b': 0})
fig.show()
```
